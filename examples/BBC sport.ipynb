```python
import texthero as hero
import pandas as pd

%load_ext autoreload
%autoreload 2

print(hero.__version__)

df = pd.read_csv(
    "https://github.com/jbesomi/texthero/raw/master/dataset/bbcsport.csv"
)

df['pca'] = (
    df['text']
    .pipe(hero.clean)
    .pipe(hero.do_tfidf)
    .pipe(hero.do_pca)
)

fig = hero.scatterplot(df, col='pca', color='topic', title="PCA BBC Sport news", return_figure=True)
fig.write_image("../github/scatterplot_bbcsport.svg")
```

```python
df['tsne'] = (
    df['text']
    .pipe(hero.clean)
    .pipe(hero.do_tfidf)
    .pipe(hero.do_tsne, perplexity=50, early_exaggeration=50)
)

#fig = hero.scatterplot(df, col='tsne', color='topic', title="TSNE BBC Sport news")
#fig.write_image("../github/scatterplot_bbcsport.svg")
```

```python
df['kmeans_labels'] = (
    df['text']
    .pipe(hero.clean)
    .pipe(hero.do_tfidf)
    .pipe(hero.do_kmeans, n_clusters=5)
)

df['kmeans_labels'] = df['kmeans_labels'].astype(str)

fig = hero.scatterplot(df, col='pca', color='kmeans_labels', title="K-means BBC Sport news", return_figure=True)
fig.write_image("../github/scatterplot_bbcsport_kmeans.svg")
```
